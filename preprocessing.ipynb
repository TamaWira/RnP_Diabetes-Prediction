```python
# mengecek outliers
outliers=[]
bmi = data['BMI']
def detect_outlier(data_1):
    
    threshold=3
    mean_1 = np.mean(data_1)
    std_1 =np.std(data_1)
    
    
    for y in data_1:
        z_score= (y - mean_1)/std_1 
        if np.abs(z_score) > threshold:
            outliers.append(y)
    return outliers


outlier_datapoints = detect_outlier(bmi)
print(outlier_datapoints)
len(outlier_datapoints)
```

```python
Q1 = np.percentile(data['BMI'], 25, interpolation = 'midpoint') 
Q2 = np.percentile(data['BMI'], 50, interpolation = 'midpoint') 
Q3 = np.percentile(data['BMI'], 75, interpolation = 'midpoint') 
  
print('Q1 25 percentile of the given data is, ', Q1)
print('Q1 50 percentile of the given data is, ', Q2)
print('Q1 75 percentile of the given data is, ', Q3)
  
IQR = Q3 - Q1 
print('Interquartile range is', IQR)
```

```python
# convert data bmi ke bentuk categorical
def bmi_convert(bmi):
    if bmi >= 30:
        return 'obesitas'

    else:
        return 'normal'

data['BMI'] = data['BMI'].apply(bmi_convert).map({'normal':0, 'obesitas':1})
data

# https://www.halodoc.com/artikel/berapa-berat-badan-yang-dikategorikan-obesitas
# https://en.wikipedia.org/wiki/Body_mass_index
```

```python
# convert data age ke bentuk categorical
def age_convert(age):
    if age >= 6:
        return 'old'

    else:
        return 'young'

data['Age'] = data['Age'].apply(age_convert).map({'young':0, 'old':1})
data

# https://sepakat.bappenas.go.id/wiki/Kelompok_Usia
# https://muamala.net/kategori-umur-menurut-who/
```
