## 1.1 Visualisasi Univariate

### a. Visualisasi Persebaran Jumlah Label 

```python
sns.countplot(x='Diabetes_binary', data=data_diabetes);
```

Dari hasil visualisasi diatas, dapat dilihat bahwa data diabetes_binary tersebut persebarannya tidak merata. Dimana data tidak diabetes (0.0) terdapat kurang lebih 190000 data dan data diabetes (1.0) kurang lebih terdapat 40000 data. Hal ini mengindikasikan bahwa dataset diabates yang digunakan memiliki kelas yang tidak seimbang. Proses oversampling dapat dilakukan untuk menyeimbangkan jumlah kelas pada dataset diabetes.

## b. Visualisasi Distribusi BMI

```python
sns.boxplot(x='BMI', data=data_diabetes);
```

```python
sns.displot(x='BMI', kde=True, data=data_diabetes);
```

Dari hasil visualisasi diatas, dapat dilihat bahwa data BMI memiliki outlier yang cukup banyak. Dan grafik dari data BMI tersebut condong ke kanan (Skewness Positif).
